### Gender Demographics
#### Percentage and Count of Male Players

```python
noDuplicatePlayers = purchase_data.drop_duplicates(subset = ['SN'])
maleCount = noDuplicatePlayers[noDuplicatePlayers.Gender == 'Male'].shape[0]
malePerc = round((maleCount/totalPlayers)*100, 2)
print(f' The Total Number of male players is {maleCount} which is {malePerc}% of the total')
```

```python
femaleCount = noDuplicatePlayers[noDuplicatePlayers.Gender == 'Female'].shape[0]
femalePerc = round((femaleCount/totalPlayers)*100, 2)
print(f' The Total Number of female players is {femaleCount} which is {femalePerc}% of the total')
```

```python
other = totalPlayers - (maleCount+femaleCount)
otherPerc = round(100 - (malePerc+femalePerc),2)
print(f' The Total Number of other players is {other} which is {otherPerc}% of the total')
```

```python
data1 = {'Gender': ['Male', 'Female', 'other/non-disclosed'],'Total Count': [maleCount, femaleCount, other], 'Percent of Players': [str(malePerc) + "%", str(femalePerc) + "%", str(otherPerc) + "%"]}
gendersummary = pd.DataFrame(data1)
gendersummary
```

### Purchasing Analysis (Gender)
#### Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

```python
MC = purchase_data[purchase_data.Gender == 'Male'].sum()
print(MC)
malePurchases = MC[6]
FC = purchase_data[purchase_data.Gender == 'Female'].sum()
femalePurchases = FC[6]
otherPurchases = totalR - malePurchases - femalePurchases
avgpricemale = round(malePurchases / totalR, 2)
avgpricefemale = round(femalePurchases / totalR, 2)
avgpriceother = round(otherPurchases / totalR, 2)
permale = round(malePurchases / maleCount,2)
perfemale = round(femalePurchases / femaleCount, 2)
perother = round(otherPurchases / other, 2)
```

```python
purchasecount = [malePurchases, femalePurchases, otherPurchases]
avgprice = [avgpricemale, avgpricefemale, avgpriceother]
perperson = [permale, perfemale, perother]
```

```python
gendersummary['Purchase Count'] = purchasecount
gendersummary['Average Price'] = avgprice
gendersummary['Average Purchase Total Per Person'] = perperson
```

### Age Demographics
#### Establish bins for ages

```python
# I did this is a really strange way I apologize, the next dataframe was done way more efficiently 
```

```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
groupnames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
noDuplicatePlayers['Age Group'] = pd.cut(noDuplicatePlayers['Age'], bins, labels=groupnames)
agedemo = noDuplicatePlayers.groupby(['Age Group']).count()
new = agedemo.drop(['Purchase ID','SN','Gender','Item ID','Item Name','Price'], axis=1)
perc = [round(x/totalPlayers*100, 2) for x in agedemo['Age']]
new['Percentage of Players'] = perc
new.rename(inplace=True,columns={'Age':'Count of Age'})
new
```
